```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
R1.plot(axs[0], 1)
R2.plot(axs[1], 1)
```

```python
#%matplotlib widget
from ipywidgets import interactive

def plot_func(index):
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(str(index-1))
    R1.plot(axs[0], index-1)
    R2.plot(axs[1], index-1)
    plt.show()
 
w = interactive(plot_func, index = (1, len(R2.traj_list)))
w
```
